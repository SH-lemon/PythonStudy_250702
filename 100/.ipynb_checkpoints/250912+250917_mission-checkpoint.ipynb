```
상세 URL(product_url)
:오른쪽을_가리키는_손_모양: 조건: 1개 카테고리를 선택하여, 각 카테고리에서 1페이지 분량만 수집합니다.
```

````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.')

driver = webdriver.Chrome(service = service, options = options)

url = "https://books.toscrape.com/catalogue/category/books/nonfiction_13/index.html"
driver.get(url)
time.sleep(3)

bookList = []

books = driver.find_elements(By.CSS_SELECTOR, "article.product_pod")
for book in books:
    title = book.find_element(By.CSS_SELECTOR, "h3>a")
    price_gbp = book.find_element(By.CSS_SELECTOR, "div.product_price>p.price_color")
    stock = book.find_element(By.CSS_SELECTOR, "p.instock.availability")
    rating = book.find_element(By.CSS_SELECTOR, "p.star-rating")
    category = "Nonfiction"
    product_url = book.find_element(By.CSS_SELECTOR, "div.image_container>a")

    rating_class = rating.get_attribute("class").split()[-1]

    bookList.append({
        "title": title.get_attribute("title").strip(),
        "price_gbp": price_gbp.text.strip(),
        "stock": stock.text.strip(),
        "rating": rating_class,
        "category": category,
        "product_url": product_url.get_attribute("href")
    })

driver.close()
print(bookList)
````

````python
import pandas as pd
table01 = pd.DataFrame(bookList)
table01
````
