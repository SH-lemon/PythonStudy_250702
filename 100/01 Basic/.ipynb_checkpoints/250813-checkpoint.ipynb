````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

options = Options()
options.add_argument("--start-maximized")
options.add_argument("window-size=1920x1080")
options.add_argument("lang=ko_KR")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://www.musinsa.com/main/musinsa/ranking?gf=A&storeCode=musinsa&sectionId=200&contentsId=&categoryCode=000&ageBand=AGE_BAND_ALL")
time.sleep(3)

links = []

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)

link_elements = driver.find_elements(By.CSS_SELECTOR, "a.sc-1t5ihy5-8.jwbFnz.gtm-select-item")
for link in link_elements :
    links.append(link.get_attribute("href").strip())

musinsa_list = []

for link in links[:50]:
    driver.get(link)
    time.sleep(3)

    title = driver.find_element(By.CSS_SELECTOR, "span.text-title_18px_med.sc-1omefes-1.exqQRL.font-pretendard").text.strip()
    brand = driver.find_element(By.CSS_SELECTOR, "span.sc-12cqkwk-2.hgzZM > span.text-body_14px_med.font-pretendard").text.strip()
    cost = driver.find_element(By.CSS_SELECTOR, "div.sc-1hw5bl8-4.EvTxQ > span.text-title_18px_semi.sc-1hw5bl8-7.kXhdZT.text-black.font-pretendard").text.strip()

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    try:
        discount = driver.find_element(By.CSS_SELECTOR, "div.sc-1hw5bl8-4.EvTxQ > span.text-title_18px_semi.sc-1hw5bl8-6.hROMjI.text-red.font-pretendard").text.strip()
    except:
        discount = "0"

    try:
        reviews = driver.find_element(By.CSS_SELECTOR, "div.GoodsReviewTitleSection__TitleContainer-sc-wkrba4-1.bKsVgj span.text-title_18px_med.font-pretendard").text.strip()
    except:
        reviews = "0"

    musinsa_list.append({
        "제품명": title,
        "브랜드": brand,
        "가격": cost,
        "할인률": discount,
        "리뷰수": reviews
    })

df = pd.DataFrame(musinsa_list)
df

driver.quit()
````

````python
from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title = "무신사 top50"

ws.append(["제품명", "브랜드", "가격", "할인률", "리뷰수"])

for m in musinsa_list :
    ws.append([m["제품명"],m["브랜드"],m["가격"],m["할인률"],m["리뷰수"]])

wb.save("musinsa_top50.xlsx")
print("저장 완료")
````
