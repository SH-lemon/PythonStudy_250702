```python
# 핏펫

import requests
import pandas as pd

def fetch_reviews(app_id, country='kr', page=1):
    url = f"https://itunes.apple.com/{country}/rss/customerreviews/page={page}/id={app_id}/sortby=mostrecent/json"
    r = requests.get(url)
    data = r.json()

    if "feed" not in data or "entry" not in data["feed"]:
        return []

    entries = data["feed"]["entry"][1:]  # 첫 번째 entry는 앱 정보
    reviews = []

    for e in entries:
        reviews.append({
            "author": e["author"]["name"]["label"],
            "rating": e["im:rating"]["label"],
            "title": e["title"]["label"],
            "content": e["content"]["label"],
            "updated": e["updated"]["label"]
        })

    return reviews

# 여러 페이지 수집
all_reviews = []
for page in range(1, 11):   # 1~10페이지 → 약 1000개
    all_reviews.extend(fetch_reviews(1335906446, 'kr', page))

df = pd.DataFrame(all_reviews)
df.head(), len(df)

# 엑셀 저장
df.to_excel("reviews.xlsx", index=False)
```
