```python
import os
import json
import zipfile
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi

# =========================
# 1️⃣ Kaggle.json 생성
# =========================
kaggle_username = "SiennaKimSH"
kaggle_key = "KGAT_738a23e3809fbe719aa88ebf44906dda"

kaggle_dir = os.path.join(os.path.expanduser("~"), ".kaggle")
os.makedirs(kaggle_dir, exist_ok=True)

kaggle_json_path = os.path.join(kaggle_dir, "kaggle.json")
with open(kaggle_json_path, "w") as f:
    json.dump({"username": kaggle_username, "key": kaggle_key}, f)

print("✅ kaggle.json 생성 완료:", kaggle_json_path)

# =========================
# 2️⃣ Kaggle API 인증
# =========================
api = KaggleApi()
api.authenticate()
print("✅ Kaggle API 인증 완료")

# =========================
# 3️⃣ Titanic 데이터 다운로드
# =========================
download_path = "titanic_data"
os.makedirs(download_path, exist_ok=True)

api.competition_download_files("titanic", path=download_path, quiet=False)
print("✅ Titanic 데이터 다운로드 완료")

# =========================
# 4️⃣ 압축 풀기
# =========================
zip_file = os.path.join(download_path, "titanic.zip")

with zipfile.ZipFile(zip_file, "r") as z:
    z.extractall(download_path)

print("✅ 압축 풀기 완료")

# =========================
# 5️⃣ CSV 파일 읽기
# =========================
train_csv = os.path.join(download_path, "train.csv")
test_csv = os.path.join(download_path, "test.csv")

train_df = pd.read_csv(train_csv)
test_df = pd.read_csv(test_csv)

print("✅ CSV 파일 로드 완료")
print("Train 데이터 샘플:")
print(train_df.head())
```

```python
import os
import requests
import zipfile
import pandas as pd

# 다운로드 폴더
os.makedirs("titanic_data", exist_ok=True)

# 파일 URL 직접 다운로드 (공개 CSV)
urls = {
    "train": "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv",
    "test": "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic_test.csv"
}

# 다운로드
for name, url in urls.items():
    r = requests.get(url)
    path = os.path.join("titanic_data", f"{name}.csv")
    with open(path, "wb") as f:
        f.write(r.content)
    print(f"✅ {name}.csv 다운로드 완료")

# CSV 읽기
train_df = pd.read_csv("titanic_data/train.csv")
test_df = pd.read_csv("titanic_data/test.csv")

print("Train 데이터 샘플:")
print(train_df.head())
```
