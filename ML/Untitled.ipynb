* 분류 : Classification : 사용자의 범주를 생성
* 회귀 : Regression : 최초 데이터를 기준으로 데이터의 변화추이 예측
* 군집화 : Clustering : 생성된 범주 가운데, 공통된 특성 및 특징을 가지고 있는 세그먼트 생성

1. KNN (K-Nearest Neighbor) => K => 매개변수
   - 머신러닝 알고리즘 > 기초모델
   - 너무 단순하기 때문에 단지 이론적으로만 이해할 뿐, 실제 사용률은 높지 않음
   - K 매개변수는 결국 데이터 담당자가 결정
   - 상대적으로 다수인 쪽의 값으로 예측하는 계획 > 정답을 맞출 수 있는 확률이 상대적으로 낮음
   - 데이터의 군집화를 해야 하는 경우에 상대적으로 많이 사용함

2. Longistic Regression (로지스틱 회귀)
   - 현실 세계에서는 "S"자 곡선 모양으로 데이터를 분류하는 경우가 많음. 그래서 이 모양을 본따서 모델 이론을 완성
   - Linear Classification 모델 : 모든 데이터를 1자 대각선을 기준으로 데이터를 분류하는 모델
   - Sigmoid function (시그모이드 함수)
