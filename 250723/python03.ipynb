```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

# 크롬 드라이버 서비스 설정
service = Service("C:\\Users\\user\\Downloads\\chromedriver\\chromedriver.exe")

# 옵션 설정
prefs = {
    "credentials_enable_service": False,
    "profile.password_manager_enabled": False
}
options = Options()
options.add_experimental_option("prefs", prefs)
options.add_argument("--disable-gpu")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

# 브라우저 실행
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://divjason.github.io/sellenium-test/")

# 로그인 입력
ID = driver.find_element(By.CSS_SELECTOR, "input#username")
PW = driver.find_element(By.CSS_SELECTOR, "input#password")
ID.clear()
PW.clear()
ID.send_keys("error@error.com")
PW.send_keys("1234")

time.sleep(5)

# 로그인 버튼 RETURN 키로 누르기
LogIn = driver.find_element(By.CSS_SELECTOR, 'input[type="submit"][value="로그인"]')
LogIn.send_keys(Keys.RETURN)

# 로그인 후 메시지 출력
message = driver.find_element(By.CSS_SELECTOR, "div.message")
print(message.text)

# 로그아웃 버튼 누르기
logOut = driver.find_element(By.CSS_SELECTOR, 'input[type="submit"][value="로그아웃"]')
logOut.send_keys(Keys.RETURN)

# 다시 로그인 페이지로 이동
driver.get("https://divjason.github.io/sellenium-test/")

# 다시 열린 페이지에서 메시지 출력
message2 = driver.find_element(By.CSS_SELECTOR, "div.message")
print(message2.text)

time.sleep(5)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

# 크롬 드라이버 서비스 설정
service = Service("C:\\Users\\user\\Downloads\\chromedriver\\chromedriver.exe")

options = Options()
options.add_experimental_option("prefs", prefs)
options.add_argument("--disable-gpu")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")

# 브라우저 실행
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://davelee-fun.github.io/")

elements = driver.find_elements(By.CSS_SELECTOR, "h4.card-text")
for element in elements :
    print(element.text)
time.sleep(3)

Next = driver.find_element(By.XPATH, '//a[text()="Next »"]')

for i in range(7) :
    Next.send_keys(Keys.RETURN)
    elements = driver.find_elements(By.CSS_SELECTOR, "h4.card-text")
    for element in elements :
        print(element.text)
time.sleep(3)
# 미완!!!!!!
driver.quit()
```
