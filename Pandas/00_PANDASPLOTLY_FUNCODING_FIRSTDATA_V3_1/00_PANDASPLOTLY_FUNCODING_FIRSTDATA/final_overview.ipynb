```python
# 1. pandas를 통해서 데이터 찾아오기

import pandas as pd

doc = pd.read_csv("00_data/olist_customers_dataset.csv", encoding="utf-8-sig")
doc.head()
```

```python
# 찾아온 데이터의 행/열 개수 확인

doc.shape
```

```python
# 찾아온 데이터의 행/열 세부정보 확인

doc.info()
```

```python
# 찾아온 데이터의 상단 필드명의 그룹을 확인하고자 할 때

doc.columns
```

```python
# 데이터의 인덱스를 확인하고자 할 때

doc.index
```

```python
# 데이터 내 중요한 5가지 지표 (5 number summary) 확인

doc.describe()
```

```python
# 기존 로우데이터에서 원하는 컬럼만 선택해서 복사본의 데이터프레임 새로 만들기

doc2 = doc[['customer_zip_code_prefix', 'customer_city', 'customer_state']].copy()

doc2
```

```python
# 복제한 데이터프레임에서 특정 컬럼 내 일치하는 값만 추출해서 개수를 확인하고자 할 때

doc3 = doc2[doc2["customer_city"] == "sao paulo"]
doc3
```

```python
doc2[["customer_city"]].drop_duplicates()
```

```python
doc2["customer_city"].unique()
```

```python
# 특정 컬럼을 그룹화한 후 나머지 컬럼의 개수를 집계할 때 사용

doc4 = doc2.groupby("customer_city").count()
doc4
```

```python
# 특정 데이터베이스 내 인덱스의 개수

len(doc4.index)
```

```python
doc4 = doc.groupby("customer_city").count()
doc4
```

```python
# 기존 데이터프레임 안에서 특정 컬럼을 기준으로 정렬하고자 할 때 (오름차순 디폴트)

doc4["customer_id"].sort_values(ascending=False).head(1)
```

```python
# 데이터프레임 안에서 특정 컬럼을 인덱스로 적용하고 싶을 때

doc.set_index("customer_city")
```
