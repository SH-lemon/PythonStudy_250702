```python
import pandas as pd

doc = pd.read_csv("COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/04-01-2020.csv", encoding="utf-8-sig")
```

```python
doc2 = doc.corr(numeric_only=True)
```

```python
import plotly.express as px

fig_px = px.imshow(
    doc2,
    # text_auto=True,
    text_auto=".2f",
    aspect = "auto",
    title = "Heatmap"
)

fig_px.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Heatmap(
        x=doc2.columns,
        y=doc2.index,
        z=doc2.values,
        colorscale="Cividis"
    )
)

fig.show()
```

```python
fig.add_trace(
    go.Heatmap(
        x=doc2.columns,
        y=doc2.index,
        z=doc2.values,
        colorscale="Viridis",
        colorbar={
            "title":"Value"
        },
        text=doc2.values,
        texttemplate="%{z:.2f}",
        hovertemplate="X : %{x}<br>Y : %{y}<br>Value : %{z:.2f}<br>Custom Info"
    )
)

fig.update_layout(
    title="Heatmap"
)

fig.show()
```
