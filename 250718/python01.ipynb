#### 현재 공유받은 크롤링증권 사이트 주식차트에서 회사명만 크롤링하여 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/blog/crawling_stock_example.html"

res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

datas = soup.select("ul.table-stock.sty1>li.row_sty")

for data in datas :
    print(data.select_one("div.st_name").get_text().strip().replace(" ",""))
```

#### 현재 크롤링해온 데이터에서 주식금액과 변동률도 함께 출력해주세요

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/blog/crawling_stock_example.html"

res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

datas = soup.select("ul.table-stock.sty1>li.row_sty")
for data in datas:
    name = data.select_one("div.st_name").get_text().strip().replace(" ","").replace("\n", "")
    price = data.select_one("div.st_price").get_text().strip().replace(" ","")
    rate = data.select_one("div.st_rate").get_text().strip().replace(" ","").replace("\n", "")
    print(f"회사명: {name}\n주가: {price}\n변동률: {rate}\n")
```
