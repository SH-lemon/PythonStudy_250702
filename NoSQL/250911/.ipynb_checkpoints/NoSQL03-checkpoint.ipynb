```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client["test"]
collection = db.daveshop

for page_number in range(10) :
    if page_number == 0 :
        res = requests.get("https://davelee-fun.github.io/")
    else :
        res = requests.get(f"https://davelee-fun.github.io/page{str(page_number+1)}")
    soup = BeautifulSoup(res.content, "html.parser")
    data = soup.select("div.card-body")
    for item in data :
        category = item.select_one("h2.card-title").get_text().replace(" 관련 상품 추천", "").strip()
        product = item.select_one("h4.card-text").get_text().replace("상품명:", "").strip()
        print(category, product)
        data = dict()
        data["category"] = category
        data["product"] = product
        collection.insert_one(data)

client.close()
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client["test"]
collection = db.daveshop

document = list(collection.find({}))
print(document[:5])
```

```python
# category 필드에 있는 모든 카테고리 종류를 조회해주세요.
categories = db.daveshop.aggregate([{"$group": {"_id": "$category"}}])

for c in categories :
    print(c)
```

```python
# 모든 상품(product)를 조회해주세요.
pipeline = ([
    {"$group": {"_id": "$product"}},
    {"$limit": 5}
])

for d in db.daveshop.aggregate(pipeline):
    print(d)
```

```python
# "매트리스커버" 카테고리의 모든 상품을 조회해주세요.
daveshop = db.daveshop
pipeline = ([
    {"$match": {"category": "매트리스커버"}}
])
for d in daveshop.aggregate(pipeline) :
    print(d)
```

```python
# 각 카테고리별 상품 수를 조회해주세요.
pipeline = ([
    {"$group": {"_id": "$category", "count": {"$sum": 1}}}
])

for d in daveshop.aggregate(pipeline) :
    print(d)
```

```python
# 상품명에 "핑크"가 포함된 상품을 조회해주세요.
pinkProduct = daveshop.find(
    {"product": {"$regex": ".*핑크.*"}}, {"_id": 0, "product": 1}
).limit(5)

for p in pinkProduct :
    print(p)
```
