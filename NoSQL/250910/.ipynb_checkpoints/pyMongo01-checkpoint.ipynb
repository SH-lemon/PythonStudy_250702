```python
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017")

db = client.sample_mflix
movies = db.movies

print(type(movies))
```

```python
for movie in movies.find({"year": 1923}, {"_id":0, "title":1, "year":1}) :
    print(movie)
```

```python
for movie in movies.find({"year": 1923}, {"_id":0, "title":1, "year":1}) :
    print(movie, movie["year"])
```

```python
# MongoDB 안에 있는 특정 DB의 컬렉션 값을 어떤 조건과 함께 찾아오게 되면 기본적으로 리스트 형태를 가진다.
# 해당 리스트 자료구조 안에 있는 각각의 문서는 딕셔너리 자료구조로 되어 있다.

for movie in movies.find(
    {"$or": [{"year": {"$lt":1900}}, {"year": {"$gt": 2015}}]},
    {"_id": 0, "title": 1, "year": 1}
) :
    print(movie)
```

```python
for movie in movies.find(
    {"genres": {"$all": ["Action", "Sci-Fi", "Drama"]}},
    {"_id":0, "genres":1, "title":1}
) :
    print(movie)
```

```python
for movie in movies.find().sort("imdb.rating", -1).skip(3).limit(3) :
    print(movie)
```

```python
# 정규표현식

import re

regex = re.compile("Star", re.I)

for movie in movies.find({"title": regex}, {"title":1}).limit(1) :
    print(movie)
```

```python
for movie in movies.find({"title": {"$regex": "Star", "$options": "i"}}, {"title":1}).limit(1) :
    print(movie)
```

```python
distinct_genres = movies.distinct("genres")
print(distinct_genres)
```

```python
# $in : 배열 중 하나라도 가졌다면 찾아옴
for movie in movies.find({"genres": {"$in": ["Action", "Adventure"]}}).limit(3) :
    print(movie)
```

```python
for movie in movies.find({"writers":{"$exists": False}}) :
    print(movie)
```

```python
count = movies.count_documents({"genres":{"$in": ["Action", "Adventure"]}})
print(count)
```
